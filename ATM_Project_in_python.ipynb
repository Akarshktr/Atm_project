```python
import time
class Atm:
    #Atm class in which all the methods are created for a particular work
    def __init__(self,name):
        self.name=name
        self.withdraw_amount=0
        self.checking_balance = 1000000
        # checking balance is a type of current balance from which you can transfer the amount in your saving balance
        self.chances=3
        self.saving_balance=100000
        self.transfer_amount=0
    def home_page(self):
        time.sleep(1.25)
        print("----------------------------------------")
        print("Pin Verified:")
        print("*************  Main Menu  *****************")
        print("")
        print(f"                      Welcome  {name}                              :")
        print(" ")
        while 1:
            print("1. Press 1 for Balance Enquiry")
            print("2. Press 2 for Deposit Money")
            print("3. Press 3 for Withdraw Money")
            print("4. Press 4 for Transfer Fund")
            print("5. Press 5 for exit The program")
            Choice = int(input())
            if Choice == 1:
                # function calling with in the class
                self.check_saving_bal()
            elif Choice == 2:
                self.deposite_funds()
            elif Choice == 3:
                self.withdraw_funds()
            elif Choice == 4:
                self.transfer_funds()
            elif Choice == 5:
                time.sleep(1.25)
                break
            else:
                print("Invalid Choice: ", '\n', "Try again Later: ")

    def check_saving_bal(self):
        # first function to check the balance
        time.sleep(1.25)    # this will give some realistic feeling by the delay of (1.25 ) secon
        print(f"Your Saving Balance is {self.saving_balance} ")
        print()
    def deposite_funds(self):
        
        # fund deposite function 
        
        time.sleep(1.25)
        deposite_amount=int(input("Enter the Amount you want To Deposit:"))
        self.saving_balance=self.saving_balance+deposite_amount
        print("Successfully Deposited")
        print(f"Total Amount after the deposition is {self.saving_balance} : ")
        print()

    def withdraw_funds(self):         # withdraw money function  
        time.sleep(1.25)
        self.withdraw_amount=int(input("Enter the amount You want to Withdraw: "))
        if self.withdraw_amount<=self.saving_balance:
            self.saving_balance=self.saving_balance-self.withdraw_amount
            print("Successfully Withdrawn: ")
            print(f"Remaining Amount after your Withdrawn from your Account is {self.saving_balance}")
        else:
            print(f"{self.withdraw_amount} Amount is not in your Account")
            print(f"Total amount in your account is {self.saving_balance}")
        print()

    def transfer_funds(self):        # transfer fund from checking_balance(current account) to saving account
        transfer_amount=int(input("Enter the value To be Transfer: "))
        if transfer_amount <= self.checking_balance:
            self.checking_balance -= transfer_amount
            self.saving_balance = self.saving_balance + transfer_amount
            self.check_saving_bal()
            # print(self.saving_balance)
            print(f"Your Current Checking balance is: {self.checking_balance} ")

        else:  # Checking if sufficient amount is not present in the checking account for transfer
            print("Not enough Balance to transfer!!\n")
            print("Current Checking balance is: ", self.checking_balance)
            print("Try with amount lesser than this balance.\n")
        print()
```
